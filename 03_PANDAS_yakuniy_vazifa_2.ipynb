Jadval ustunlari ko'mir, gaz, gidro va boshqa manbaalardan ko'rsatilgan yilda qancha __terravat/soat__ energiya olinganini ko'rsatadi.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/Electricity_Production_By_Source.csv")
df.columns = ['davlat','kod','yil','komir','gaz','gidro','yangilanuvchi','quyosh','neft','shamol','atom']
df.head()
```

### 5-VAZIFA. AQSh 2020-yilda jami qancha elektr energiyasi ishlab chiqqan?

```python
# JAVOBNI SHUYERGA YOZING
aqw = pd.Series(dfus.loc[2020][['komir','gaz','gidro','yangilanuvchi','quyosh','neft','shamol','atom']].sum(),index = [2020])
aqw
```

### 6-VAZIFA. 2010-2020-yillarda AQSH turli manbaalardan o'rtacha qancha elektr toki olganini chiqaring. Natijani kichikdan kattaga qarab tartiblang.

```python
# JAVOBNI SHUYERGA YOZING
dfus = df.loc[df['davlat']=='United States']
ortacha = dfus.loc[(dfus['yil']>=2010)&(dfus['yil']<=2020),['komir','gaz','gidro','yangilanuvchi','quyosh','neft','shamol','atom']].mean(axis=0).sort_values()
ortacha
```

```python
#@title O'xshash natija (1990-2000 yillar uchun)
```

### 7-VAZIFA. AQSH gidro elektr stansiyalardan eng kam va eng ko'p elektr energiyasi olgan yillarni toping.

```python
# JAVOBNI SHUYERGA YOZING
eng_kam_gidro_yil = df[df['gidro'] == df['gidro'].min()]['yil'].values[0]
eng_kam_gidro_yil
```

```python
eng_kop_gidro_yil = df[df['gidro']==df['gidro'].max()]['yil'].values[0]
eng_kop_gidro_yil
```

### 10-VAZIFA.
#### 10.1. Asl DF dan 2000-2020-yillar, `France` uchun `atom`, `komir`, `gaz`, `quyosh` manbaalarini alohida `dffr` jadvaliga ko'chirib oling
#### 10.2. `dffr` jadvalini yil ustuni bo'yicha tartiblang

```python
# JAVOBNI SHUYERGA YOZING
fr = df.loc[df['davlat']=='France',['yil','atom','komir','gaz','quyosh']]
dffr = fr[(fr['yil']>=2000)&(fr['yil']<=2020)]
dffr.sort_index(axis = 1)
```
